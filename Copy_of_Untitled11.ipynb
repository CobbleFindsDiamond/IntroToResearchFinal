```python
import os
import math
import random
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import train_test_split

from torchvision import transforms
from PIL import Image
import timm

# ============================================================
# Config
# ============================================================

CSV_PATH_A = "/content/data_labels.csv"        # Dataset 1
CSV_PATH_B = "/content/data_labels_diverse.csv"  # Dataset 2 (change this)

BATCH_SIZE = 32
EPOCHS = 50
PATIENCE = 7          # early stopping patience
LR = 1e-4             # learning rate
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# Set seeds for reproducibility (optional)
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if DEVICE == "cuda":
    torch.cuda.manual_seed_all(SEED)

label_cols = ["kcal", "protein", "fat", "carbs"]

# ============================================================
# Dataset
# ============================================================

class FoodDataset(Dataset):
    def __init__(self, df: pd.DataFrame, transform=None):
        self.df = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        img_path = row["image_path"]
        img = Image.open(img_path).convert("RGB")

        if self.transform:
            img = self.transform(img)

        target = torch.tensor([
            row["kcal"],
            row["protein"],
            row["fat"],
            row["carbs"]
        ], dtype=torch.float32)

        return img, target

# ============================================================
# Transforms (shared across datasets)
# ============================================================

train_tfms = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

val_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# ============================================================
# Model definition utilities
# ============================================================

class NutritionRegressor(nn.Module):
    def __init__(self, backbone, embed_dim):
        super().__init__()
        self.backbone = backbone
        self.fc = nn.Sequential(
            nn.Linear(embed_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 4)   # standardized nutrition outputs
        )

    def forward(self, x):
        feats = self.backbone(x)
        out = self.fc(feats)
        return out

def create_backbone_and_dim():
    backbone = timm.create_model("efficientnet_b0", pretrained=True)
    backbone.classifier = nn.Identity()   # remove original classification head

    with torch.no_grad():
        dummy = torch.randn(1, 3, 224, 224)
        feat_dim = backbone(dummy).shape[1]

    # Freeze everything first
    for p in backbone.parameters():
        p.requires_grad = False

    # Unfreeze last two blocks + head
    if hasattr(backbone, "blocks"):
        for p in backbone.blocks[-2:].parameters():
            p.requires_grad = True

    if hasattr(backbone, "conv_head"):
        for p in backbone.conv_head.parameters():
            p.requires_grad = True
    if hasattr(backbone, "bn2"):
        for p in backbone.bn2.parameters():
            p.requires_grad = True

    return backbone, feat_dim

# ============================================================
# Loss
# ============================================================

l1 = nn.L1Loss()
mse = nn.MSELoss()

def regression_loss(pred, target_std):
    return 0.5 * l1(pred, target_std) + 0.5 * mse(pred, target_std)

# ============================================================
# Training routine for ONE CSV
# ============================================================

def train_on_csv(csv_path, tag="dataset"):
    print(f"\n==============================")
    print(f"Training model on {csv_path} ({tag})")
    print(f"==============================\n")

    # ----- Load CSV & split -----
    df = pd.read_csv(csv_path)

    train_df, val_df = train_test_split(
        df,
        test_size=0.2,
        random_state=SEED,
        shuffle=True
    )

    print("Train samples:", len(train_df))
    print("Val samples:  ", len(val_df))

    # Label stats (TRAIN ONLY) for this dataset
    label_mean = train_df[label_cols].mean().values.astype(np.float32)
    label_std  = train_df[label_cols].std().values.astype(np.float32)
    label_std[label_std == 0] = 1.0  # avoid div by zero

    print("\nLabel mean:", label_mean)
    print("Label std: ", label_std)

    label_mean_t = torch.tensor(label_mean, device=DEVICE)
    label_std_t  = torch.tensor(label_std,  device=DEVICE)

    # ----- Datasets & loaders -----
    train_dataset = FoodDataset(train_df, transform=train_tfms)
    val_dataset   = FoodDataset(val_df,   transform=val_tfms)

    train_loader = DataLoader(
        train_dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=2,
        pin_memory=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    # ----- Model, optimizer, scheduler -----
    backbone, feat_dim = create_backbone_and_dim()
    backbone = backbone.to(DEVICE)

    model = NutritionRegressor(backbone, feat_dim).to(DEVICE)

    optimizer = optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=LR
    )

    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer,
        mode="min",
        factor=0.5,
        patience=3
    )

    # ----- Training loop with early stopping -----
    best_val_loss = float("inf")
    best_state = None
    epochs_no_improve = 0

    for epoch in range(1, EPOCHS + 1):
        # -------------------------
        # Training
        # -------------------------
        model.train()
        total_train_loss = 0.0

        for imgs, targets in train_loader:
            imgs = imgs.to(DEVICE)
            targets = targets.to(DEVICE)

            targets_std = (targets - label_mean_t) / label_std_t

            preds_std = model(imgs)
            loss = regression_loss(preds_std, targets_std)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item()

        avg_train_loss = total_train_loss / len(train_loader)

        # -------------------------
        # Validation
        # -------------------------
        model.eval()
        total_val_loss = 0.0
        all_val_abs_err = []

        with torch.no_grad():
            for imgs, targets in val_loader:
                imgs = imgs.to(DEVICE)
                targets = targets.to(DEVICE)

                targets_std = (targets - label_mean_t) / label_std_t
                preds_std = model(imgs)

                loss = regression_loss(preds_std, targets_std)
                total_val_loss += loss.item()

                preds = preds_std * label_std_t + label_mean_t
                abs_err = torch.abs(preds - targets)
                all_val_abs_err.append(abs_err.cpu())

        avg_val_loss = total_val_loss / len(val_loader)

        all_val_abs_err = torch.cat(all_val_abs_err, dim=0)
        mae_cal = all_val_abs_err[:, 0].mean().item()
        mae_pro = all_val_abs_err[:, 1].mean().item()
        mae_fat = all_val_abs_err[:, 2].mean().item()
        mae_carb = all_val_abs_err[:, 3].mean().item()
        overall_mae = all_val_abs_err.mean().item()

        print(f"\n[{tag}] Epoch {epoch}/{EPOCHS}")
        print(f"Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
        print(f"Val MAE (kcal, g):")
        print(f"  Calories: {mae_cal:.3f}")
        print(f"  Protein : {mae_pro:.3f}")
        print(f"  Fat     : {mae_fat:.3f}")
        print(f"  Carbs   : {mae_carb:.3f}")
        print(f"  Overall MAE: {overall_mae:.3f}")

        scheduler.step(avg_val_loss)

        # Early stopping
        if avg_val_loss < best_val_loss - 1e-4:
            best_val_loss = avg_val_loss
            best_state = {
                "model": model.state_dict(),
                "optimizer": optimizer.state_dict(),
                "epoch": epoch,
                "label_mean": label_mean,
                "label_std": label_std,
            }
            epochs_no_improve = 0
            print("  -> New best model saved.")
        else:
            epochs_no_improve += 1
            print(f"  -> No improvement for {epochs_no_improve} epoch(s).")

        if epochs_no_improve >= PATIENCE:
            print(f"\n[{tag}] Early stopping triggered.")
            break

    print(f"\n[{tag}] Training finished.")

    if best_state is not None:
        model.load_state_dict(best_state["model"])

    # Return everything you might want later
    return {
        "model": model,
        "best_state": best_state,
        "label_mean": label_mean,
        "label_std": label_std,
        "best_val_loss": best_val_loss,
        "csv_path": csv_path,
        "tag": tag,
    }

# ============================================================
# Train TWO models: one per CSV
# ============================================================

results_a = train_on_csv(CSV_PATH_A, tag="dataset_A")
results_b = train_on_csv(CSV_PATH_B, tag="dataset_B")

# Example: access models / stats
model_a = results_a["model"]
model_b = results_b["model"]

print("\nDone training both models.")
print("Best val loss A:", results_a["best_val_loss"])
print("Best val loss B:", results_b["best_val_loss"])
```

```python
# ============================================================
# STEP 7 â€” Compare BOTH models on the SAME Diverse Test Set
# ============================================================

print("\n==============================================")
print("EVALUATING BOTH MODELS ON DIVERSE DATASET ONLY")
print("==============================================")

# We'll reuse Dataset B as the DIVERSE test dataset
diverse_df = pd.read_csv(CSV_PATH_B)

# You can optionally resplit just for testing if needed
_, diverse_test_df = train_test_split(
    diverse_df,
    test_size=0.2,
    random_state=SEED,
    shuffle=True
)

diverse_test_dataset = FoodDataset(diverse_test_df, transform=val_tfms)

diverse_test_loader = DataLoader(
    diverse_test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# Unpack models + normalization stats
model_A = results_a["model"]
model_B = results_b["model"]

mean_A = torch.tensor(results_a["label_mean"], device=DEVICE)
std_A  = torch.tensor(results_a["label_std"],  device=DEVICE)

mean_B = torch.tensor(results_b["label_mean"], device=DEVICE)
std_B  = torch.tensor(results_b["label_std"],  device=DEVICE)


def evaluate_model_on_loader(model, loader, mean_t, std_t):
    model.eval()
    abs_errs = []

    with torch.no_grad():
        for imgs, targets in loader:
            imgs = imgs.to(DEVICE)
            targets = targets.to(DEVICE)

            preds_std = model(imgs)
            preds = preds_std * std_t + mean_t

            abs_err = torch.abs(preds - targets)
            abs_errs.append(abs_err.cpu())

    abs_errs = torch.cat(abs_errs, dim=0)

    return {
        "kcal": abs_errs[:, 0].mean().item(),
        "protein": abs_errs[:, 1].mean().item(),
        "fat": abs_errs[:, 2].mean().item(),
        "carbs": abs_errs[:, 3].mean().item(),
        "overall": abs_errs.mean().item()
    }


# ---- Evaluate BOTH models on the SAME diverse test set ----

mae_A = evaluate_model_on_loader(model_A, diverse_test_loader, mean_A, std_A)
mae_B = evaluate_model_on_loader(model_B, diverse_test_loader, mean_B, std_B)

# ---- Print Clean Comparison Table ----

print("\n================ MAE COMPARISON (DIVERSE TEST SET) ================")
print(f"{'Metric':<10} | {'Model A MAE':>12} | {'Model B MAE':>12} | {'Î” (A - B)':>12}")
print("-" * 60)

for k in ["kcal", "protein", "fat", "carbs", "overall"]:
    diff = mae_A[k] - mae_B[k]
    print(f"{k:<10} | {mae_A[k]:>12.3f} | {mae_B[k]:>12.3f} | {diff:>+12.3f}")

print("===================================================================")

# Interpretation helper printout
print("\nINTERPRETATION:")
print(" â€¢ Negative Î”  â†’ Model A is better (lower error)")
print(" â€¢ Positive Î”  â†’ Model B is better (lower error)")
print(" â€¢ 'overall'   â†’ Mean of all nutrition errors")
```

```python
# ============================================================
# STEP 8 â€” CLASSIFIER-STYLE CONSTANT BASELINE COMPARISON
# (Evaluated on the SAME Diverse Test Set)
# ============================================================

print("\n==============================================")
print("CLASSIFIER-STYLE BASELINE vs MODELS (DIVERSE)")
print("==============================================")

# ---- True values from Diverse Test Set ----
y_true = diverse_test_df[["kcal", "protein", "fat", "carbs"]].values

# ---- Constant Baseline (Mean Predictor) ----
baseline_vec = y_true.mean(axis=0)
y_pred_baseline = np.tile(baseline_vec, (len(y_true), 1))

# ---- Baseline Absolute Error ----
abs_err_base = np.abs(y_pred_baseline - y_true)

cal_MAE_base     = abs_err_base[:, 0].mean()
protein_MAE_base = abs_err_base[:, 1].mean()
fat_MAE_base     = abs_err_base[:, 2].mean()
carb_MAE_base    = abs_err_base[:, 3].mean()
overall_MAE_base = abs_err_base.mean()

print("\n=== Baseline (Constant / Classifier-Style) MAE ===")
print(f"Calories MAE: {cal_MAE_base:.3f}")
print(f"Protein MAE:  {protein_MAE_base:.3f}")
print(f"Fat MAE:      {fat_MAE_base:.3f}")
print(f"Carbs MAE:    {carb_MAE_base:.3f}")
print(f"Overall MAE:  {overall_MAE_base:.3f}")


# ============================================================
# Compare Baseline vs Model A vs Model B
# ============================================================

print("\n================ FINAL COMPARISON TABLE =================")
print(f"{'Metric':<10} | {'Baseline':>10} | {'Model A':>10} | {'Model B':>10}")
print("-" * 60)

print(f"{'kcal':<10} | {cal_MAE_base:>10.3f} | {mae_A['kcal']:>10.3f} | {mae_B['kcal']:>10.3f}")
print(f"{'protein':<10} | {protein_MAE_base:>10.3f} | {mae_A['protein']:>10.3f} | {mae_B['protein']:>10.3f}")
print(f"{'fat':<10} | {fat_MAE_base:>10.3f} | {mae_A['fat']:>10.3f} | {mae_B['fat']:>10.3f}")
print(f"{'carbs':<10} | {carb_MAE_base:>10.3f} | {mae_A['carbs']:>10.3f} | {mae_B['carbs']:>10.3f}")
print(f"{'overall':<10} | {overall_MAE_base:>10.3f} | {mae_A['overall']:>10.3f} | {mae_B['overall']:>10.3f}")

print("=========================================================")

# ============================================================
# Generalization Gain vs Baseline
# ============================================================

print("\n=== GENERALIZATION GAIN vs BASELINE (LOWER IS BETTER) ===")
print(f"Model A improvement over baseline: {overall_MAE_base - mae_A['overall']:.3f}")
print(f"Model B improvement over baseline: {overall_MAE_base - mae_B['overall']:.3f}")
```
